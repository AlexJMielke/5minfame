# Data Visualisation Catalogue

https://datavizcatalogue.com/index.html

very nice overview for various types of data visualization including:
- comparisons
- proportions
- relationships
- hierarchy
- concepts
- location
- part-to-a-whole
- distribution
- how things work
- processes & methods
- movement or flow
- patterns
- range
- data over time
- analysing text
- reference tool

every entry comes with:
- discription
- anatomy
- function
- similar charts
- tools to generate visualisation
- examples

Can be used as a source of inspiration for your data visualisation
